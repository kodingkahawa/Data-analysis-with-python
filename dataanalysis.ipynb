```python
# Creating Datafame from random values

from random import randint

# Declare the columns that we will need 
columns = ['a', 'b', 'c']

df = pd.DataFrame(columns=columns)


# The .loc method works in the same way slicing works in python list,loc[] to insert the random number in either of the columns 
# thus we are simply updating the null elements with values.
for i in range(7):
    for c in columns:
      df.loc[i,c] = randint(-1,1)
  
# and printing out the dataframe
df
```

### Working on missing data

#### Fillna methods

```python
Fill_no = bike.fillna(1000000)  # fill in missing with 1000000 and save the data to Fill_no
Fill_no.isnull().values.any()
```

```python

Fill_str=bike["TO_ST"].fillna("missing")  # fill in missing with a string: "missing" and save the data to Fill_str
Fill_str.isnull().sum()
```

```python

bike["SHAPE_LEN"].fillna(bike["SHAPE_LEN"].mean(), inplace=True)  # fill missing values with the sample mean and save the changes to the original data
```

```python
bike["SHAPE_LEN"].isnull().values.any()
```

#### Other methods

```python
bike.drop_duplicates(inplace=True)  # remove duplicates, if any, and save the changes to the original data
```

```python
subset1 = bike.drop(bike.index[[1,7,9]])  # drop the 2nd, 8th, and 10th rows
```

```python
subset2 = bike.drop(bike.index[range(1,11)])  # drop all rows between 2nd to 10th rows
```
