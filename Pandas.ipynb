```python
# Let's import the pandas library 
# 
import pandas as pd
```

```python
# Converting dictionaries to dataframes
data = [{'name': 'vikash', 'age': 27}, {'name': 'Satyam', 'age': 14}]
df = pd.DataFrame.from_dict(data, orient='columns')
df
```

```python
# Creating Datafame from random values

from random import randint

# Declare the columns that we will need 
columns = ['a', 'b', 'c']

df = pd.DataFrame(columns=columns)


# The .loc method works in the same way slicing works in python list,loc[] to insert the random number in either of the columns 
# thus we are simply updating the null elements with values.
for i in range(7):
    for c in columns:
      df.loc[i,c] = randint(-1,1)
  
# and printing out the dataframe
df
```

```python
# Changing Dataframe column names 
# 
student_list = [['Derrick', 1, 'a'],['Austin', 2, 'a'],['Dorris', 3, 'a']]

students = pd.DataFrame(student_list, columns = ['name','class','grade'])
students

students.columns.values[1:] = ['student_' + val for val in students.columns.values[1:]]
students.columns.values
students
```

```python
# A simpler way of changing Dataframe column names 
#
student_list = [['Derrick', 1, 'a'],['Austin', 2, 'a'],['Dorris', 3, 'a']]
students = pd.DataFrame(student_list, columns = ['name','class','grade'])
students

# Uncomment the following line
students.columns = ['names', 'Grade', 'Score']
students
```

```python
# A simpler way of changing Dataframe column names 
#
student_list = [['Derrick', 1],['Austin', 2],['Dorris', 3]]
students = pd.DataFrame(student_list, columns = ['name','class',])


students_list2 = [['Derrick', 'A'],['Austin',"B" ],['Dorris',"A"]]
students1 = pd.DataFrame(students_list2, columns = ['name','grade',])
```
